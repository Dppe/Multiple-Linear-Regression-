```python
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Importing the dataset
dt = pd.read_csv(r'50_Startups.csv')
dt
```

```python
dt.head()
```

```python
dt.info()
```

```python
dt.describe()
```

```python
x= dt.iloc[:, :-1]
y= dt.iloc[:, 4]
```

```python
x
```

```python
y
```

```python
#Convert the column into categorical columns
states=pd.get_dummies(x['State'],drop_first=False)
states
```

```python
# Drop the state coulmn
X=x.drop('State',axis=1)
```

```python
X
```

```python
# concat the dummy variables
X=pd.concat([X,states],axis=1)
```

```python
X
```

```python
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
X_train
```

```python
X_test
```

```python
y_train
```

```python
y_test
```

```python
# Fitting Multiple Linear Regression to the Training set
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
# Predicting the Test set results
y_pred = regressor.predict(X_test)
y_pred
```

```python
from sklearn.metrics import r2_score
score=r2_score(y_test,y_pred)
```

```python
score=score*100
score
```

```python
# Visualizing the results

# Plotting the Actual vs Predicted values
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred, color='blue')
plt.plot(y_test, y_test, color='red')  # Line representing perfect predictions
plt.title('Actual vs Predicted values')
plt.xlabel('Actual values')
plt.ylabel('Predicted values')
plt.grid(True)
plt.show()

```

```python
# Plotting the residuals
residuals = y_test - y_pred
plt.figure(figsize=(10, 6))
plt.scatter(y_pred, residuals, color='purple')
plt.hlines(y=0, xmin=y_pred.min(), xmax=y_pred.max(), color='red')
plt.title('Residuals Plot')
plt.xlabel('Predicted values')
plt.ylabel('Residuals')
plt.grid(True)
plt.show()
```
